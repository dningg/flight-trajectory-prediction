```python
import pandas as pd
```

```python
data = pd.read_csv('combined_csv_data.csv')
```

```python
data.info()
```

```python
def extract_first_percentage_rows(percentage):
    num_rows = int(len(data) * (percentage / 100))
    first_percentage_rows = data.head(num_rows)
    first_percentage_rows.to_csv('first10%_data.csv', index=False)

# Gọi hàm với phần trăm cụ thể
extract_first_percentage_rows(10)
```
